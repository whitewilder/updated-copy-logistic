```python
import pandas as pd
import numpy as np

class Transformation:
    
    def __init__(self, df):
        self.df = df.copy()
        self.encoders = {}  # Dictionary to store encoders

    def _check_column_exists(self, col):
        if col not in self.df.columns:
            raise ValueError(f"Column '{col}' does not exist in the DataFrame.")
    
    def _fill_missing_years(self, df, id_col, date_col, value_col):
        df = df.copy()
        df[date_col] = pd.to_datetime(df[date_col])
        df = df.sort_values(by=[id_col, date_col])
        
        # Create a DataFrame to hold the filled data
        filled_df = pd.DataFrame()
        
        for cust_id, group in df.groupby(id_col):
            group = group.set_index(date_col)
            all_years = pd.date_range(start=group.index.min(), end=group.index.max(), freq='YS')
            full_group = group.reindex(all_years)
            
            # Identify large gaps and fill only if the gap is more than 2 years
            missing_years = full_group[full_group[value_col].isna()].index.year
            previous_years = full_group[full_group[value_col].notna()].index.year
            gaps = missing_years.difference(previous_years)
            
            for year in gaps:
                if any((previous_years < year - 1) & (previous_years > year - 3)):
                    full_group.at[pd.Timestamp(f"{year}-01-01"), value_col] = np.nan
            
            # Interpolate missing values
            full_group[value_col] = full_group[value_col].interpolate(method='linear')
            
            full_group[id_col] = cust_id
            full_group = full_group.reset_index().rename(columns={'index': date_col})
            filled_df = pd.concat([filled_df, full_group])
        
        return filled_df

    def calculate_growth(self, id_col, date_col, value_col, suffix='_growth'):
        self._check_column_exists(id_col)
        self._check_column_exists(date_col)
        self._check_column_exists(value_col)

        self.df[date_col] = pd.to_datetime(self.df[date_col])
        self.df = self.df.sort_values(by=[id_col, date_col])
        
        # Fill missing years
        self.df = self._fill_missing_years(self.df, id_col, date_col, value_col)
        
        # Calculate growth
        self.df[f'{value_col}{suffix}'] = self.df.groupby(id_col)[value_col].pct_change() * 100
        return self.df

# Example usage:
# Create a sample DataFrame
data = {
    'cust_id': [1, 1, 1, 2, 2, 3, 3],
    'date': ['2020-01-01', '2021-01-01', '2023-01-01', '2020-01-01', '2022-01-01', '2020-01-01', '2021-01-01'],
    'turnover': [100, 110, 130, 200, 220, 300, 330]
}
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])

# Instantiate the Transformation class with the DataFrame
transformer = Transformation(df)

# Calculate turnover growth
transformed_df = transformer.calculate_growth(id_col='cust_id', date_col='date', value_col='turnover')

print(transformed_df)
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import roc_auc_score

def gini_coefficient(y_true, y_pred):
    auc = roc_auc_score(y_true, y_pred)
    return 2 * auc - 1

def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Variable"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

def forward_selection(X, y, scoring='gini'):
    remaining_features = X.columns.tolist()
    selected_features = []
    best_score = -np.inf
    
    while remaining_features:
        feature_scores = []
        for feature in remaining_features:
            features = selected_features + [feature]
            X_train_with_features = X[features].copy()
            X_train_with_features = sm.add_constant(X_train_with_features)
            
            try:
                model = sm.Logit(y, X_train_with_features).fit(disp=0)
                y_pred_prob = model.predict(X_train_with_features)
                
                if scoring == 'gini':
                    score = gini_coefficient(y, y_pred_prob)
                elif scoring == 'pvalue':
                    score = model.pvalues[feature]
                
                feature_scores.append((feature, score))
            
            except np.linalg.LinAlgError:
                print(f"Skipping feature {feature} due to singular matrix error.")
                continue
        
        if not feature_scores:
            break
        
        feature_scores.sort(key=lambda x: x[1], reverse=(scoring == 'gini'))
        best_feature, best_score_new = feature_scores[0]
        
        if best_score_new > best_score or (scoring == 'pvalue' and best_score_new < 0.05):
            selected_features.append(best_feature)
            remaining_features.remove(best_feature)
            best_score = best_score_new
        else:
            break
    
    return selected_features

def backward_selection(X, y, scoring='gini'):
    features = X.columns.tolist()
    best_score = -np.inf
    
    while len(features) > 0:
        X_train_with_features = X[features].copy()
        X_train_with_features = sm.add_constant(X_train_with_features)
        
        try:
            model = sm.Logit(y, X_train_with_features).fit(disp=0)
            y_pred_prob = model.predict(X_train_with_features)
            
            if scoring == 'gini':
                current_score = gini_coefficient(y, y_pred_prob)
            elif scoring == 'pvalue':
                p_values = model.pvalues[1:]  # Exclude intercept
                current_score = p_values.max()
            
            if current_score > best_score or (scoring == 'pvalue' and current_score < 0.05):
                best_score = current_score
            else:
                # Remove the least significant feature
                p_values = model.pvalues[1:]  # Exclude intercept
                worst_feature = p_values.idxmax()
                features.remove(worst_feature)
        
        except np.linalg.LinAlgError:
            print(f"Skipping iteration due to singular matrix error.")
            break
    
    return features

def model_stats(X, y, features):
    X_train_with_features = X[features].copy()
    X_train_with_features = sm.add_constant(X_train_with_features)
    
    try:
        model = sm.Logit(y, X_train_with_features).fit()
        print(model.summary())
        
        vif_data = calculate_vif(X_train_with_features.iloc[:, 1:])  # Exclude the constant
        print("\nVariance Inflation Factor (VIF):")
        print(vif_data)
    
    except np.linalg.LinAlgError:
        print("Model fitting failed due to singular matrix error.")

def feature_selection_pipeline(X, y, method='forward', scoring='gini'):
    # Check for constant columns or columns with zero variance
    if X.var().min() == 0:
        constant_cols = X.columns[X.var() == 0].tolist()
        if constant_cols:
            print(f"Removing constant columns: {constant_cols}")
            X = X.drop(columns=constant_cols)
    
    # Standardize features
    scaler = StandardScaler()
    X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns, index=X.index)
    
    # Ensure y has the same index as X
    y = y.loc[X.index]
    
    if method == 'forward':
        selected_features = forward_selection(X, y, scoring)
    elif method == 'backward':
        selected_features = backward_selection(X, y, scoring)
    else:
        raise ValueError("Method must be 'forward' or 'backward'")
    
    print(f"Selected features ({method.capitalize()} Selection with {scoring.capitalize()} scoring):", selected_features)
    model_stats(X, y, selected_features)

# Example usage
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Run forward selection
feature_selection_pipeline(X_train, y_train, method='forward', scoring='gini')

# Run backward selection
feature_selection_pipeline(X_train, y_train, method='backward', scoring='pvalue')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x, c):
    return 1 / (1 + np.exp(-(x + c)))

# Define range for x
x = np.linspace(-30, 30, 400)

# Define a fixed constant
constant = 2

# Generate multiple sets of x values
x_sets = [x, x + 10.5, x - 50.5]  # Different shifts in x for demonstration

# Plot sigmoid function for varying x sets with the same constant
plt.figure(figsize=(12, 8))

for i, x_set in enumerate(x_sets):
    y = sigmoid(x_set, constant)
    plt.plot(x_set, y, label=f'X set {i + 1}')

plt.title('Sigmoid Function with Varying x and Fixed Constant')
plt.xlabel('x')
plt.ylabel('Sigmoid(x + c)')
plt.legend()
plt.grid(True)
plt.show()
```
